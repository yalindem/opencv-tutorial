```python
img_bgr = cv2.imread('../DATA/00-puppy.jpg')
plt.imshow(img_bgr)
```

Das Bild wurde von OpenCV korrekt als NumPy-Array geladen, aber die Farbe jedes Pixels wurde als BGR sortiert. Matplotlib's plot erwartet ein RGB-Bild. Daher ist es für die korrekte Darstellung des Bildes nötig, diese Kanäle zu vertauschen. Diese Operation kann entweder mit der OpenCV-Konvertierungsfunktion cv2.cvtColor() oder direkt auf dem NumPy-Array ausgeführt werden.

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('../DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
img_gray = cv2.imread('../DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray,cmap='gray')
```

## Bildgrösse anpassen

```python
img_rgb.shape
# Breite, Höhe, Farbkanäle
```

```python
img =cv2.resize(img_rgb,(1300,275))
```

```python
plt.imshow(img)
```

### Nach Verhältnis

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img =cv2.resize(img_rgb,(0,0),img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
%matplotlib qt
plt.imshow(new_img)
```

### Bilder spiegeln

```python
%matplotlib inline
```

```python
# Entlang der zentralen X-Achse
new_img = cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
# Entlang der zentralen Y-Achse
new_img = cv2.flip(new_img,1)
plt.imshow(new_img)
```

```python
# Entlang beider Achsen
new_img = cv2.flip(new_img,-1)
plt.imshow(new_img)
```
