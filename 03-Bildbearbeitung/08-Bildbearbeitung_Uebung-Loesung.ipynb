**Einiger nützlicher Code ist bereits hier für dich in den folgenden Zellen:**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

**AUFGABE: Öffne das im Verzeichnis DATA befindliche Bild giraffes.jpg und zeige es hier an.**

```python
img = cv2.imread('../DATA/giraffes.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display_img(img)
```

**Aufgabe:Wende einen binären Schwellenwert (binary threshold) auf das Bild an.**

```python
img = cv2.imread('../DATA/giraffes.jpg',0)
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
display_img(thresh1,cmap='gray')
```

**Aufgabe: Öffne giraffes.jpg aus dem Verzeichnis DATA, konvertiere den Farbraum zu HSV und zeige das Bild an.**

```python
img = cv2.imread('../DATA/giraffes.jpg')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
display_img(img)
```

**AUFGABE: Erzeuge einen Tiefenpassfilter mit einem 4 mal 4 Kernel, befüllt mit Werten von 1/10 (0.01) und verwende dann 2D-Faltung (2D convolution), um die Unschärfe des Giraffenbildes zu erhöhen (Anzeige in normaler RGB-Reihenfolge)**

```python
kernel = np.ones(shape=(4,4),dtype=np.float32)/10
```

```python
kernel
```

```python
img = cv2.imread('../DATA/giraffes.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
dst = cv2.filter2D(img,-1,kernel)
display_img(dst)
```

**AUFGABE: Erzeuge einen horizontalen Sobel-Filter (sobelx aus unserer Lektion) mit einer Kernelgrösse von 5 für die Graustufenversion der Giraffen und zeige  dann die nach resultierendem Gradienten gefiltere Version des Bildes an.**

```python
img = cv2.imread('../DATA/giraffes.jpg',0)
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
```

```python
display_img(sobelx,cmap='gray')
```

**AUFGABE: Plotte das Farbhistogramm für den Rot- (RED), Blau- (BLUE) und Grün-Kanal (GREEN) des Giraffenbildes. Bitte beachte die Reihenfolge der Kanäle hierbei genau.**

```python
img =cv2.imread('../DATA/giraffes.jpg')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('Giaraffes Histograms')
plt.show()
```
