# Farbräume (Color Spaces)

Lasst uns eine kurze Einführung in die Konvertierung verschiedener Farbräume nachvollziehen! Das Video enthält mehr Details zu Farbräumen.

Link mit weiterführenden Informationen: 
https://en.wikipedia.org/wiki/HSL_and_HSV


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread('../DATA/00-puppy.jpg')
```

### Konvertierung in verschiedene Farbräume

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

**Konvertierung nach HSV**

https://en.wikipedia.org/wiki/HSL_and_HSV

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img)
```

```python
img = cv2.imread('../DATA/00-puppy.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
plt.imshow(img)
```
