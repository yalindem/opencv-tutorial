# Morphologische Operatoren (Morphological Operators)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img =np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text='ABCDE',org=(50,300), fontFace=font,fontScale= 5,color=(255,255,255),thickness=25,lineType=cv2.LINE_AA)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = load_img()
```

```python
display_img(img)
```

## Erosion

Begrenzungen von Vordergründen werden wegerodiert. Funktioniert am besten wenn der Vordergrund eine helle Farbe (vorzugsweise Weiß) hat und der Hintergrund dunkel ist.

```python
kernel = np.ones((5,5),np.uint8)
erosion1 = cv2.erode(img,kernel,iterations = 1)
```

```python
display_img(erosion1)
```

```python
img = load_img()
kernel = np.ones((5,5),np.uint8)
erosion5 = cv2.erode(img,kernel,iterations = 4)
```

```python
display_img(erosion5)
```

## Öffnung (Opening)
Öffnung ist Erosion gefolgt von einer Ausdehnung (dilation). Nützlich zur Entfernung von Störsignalen im Hintergrund!

```python
img = load_img()
```

```python
white_noise = np.random.randint(low=0,high=2,size=(600,600))
```

```python
white_noise
```

```python
white_noise = white_noise*255
```

```python
white_noise.shape
```

```python
img.shape
```

```python
noise_img = white_noise+img
```

```python
display_img(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel)
```

```python
display_img(opening)
```

### Schließung (Closing)
Nützlich bei der Entfernung von Störsignalen von Vordergrund-Objekten, wie schwarzen Punkten auf einem weissen Text.

```python
img = load_img()
```

```python
black_noise = np.random.randint(low=0,high=2,size=(600,600))
```

```python
black_noise
```

```python
black_noise= black_noise * -255
```

```python
black_noise_img = img + black_noise
```

```python
black_noise_img
```

```python
black_noise_img[black_noise_img==-255] = 0
```

```python
display_img(black_noise_img)
```

```python
closing = cv2.morphologyEx(black_noise_img, cv2.MORPH_CLOSE, kernel)
```

```python
display_img(closing)
```

## Morphologischer Gradient

Die Differenz zwischen der Ausdehnung und Erosion eines Bildes.

```python
img = load_img()
```

```python
display_img(img)
```

```python
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
```

```python
display_img(gradient)
```
